```python
# 10.

import math
import numpy

hipp['D'] = (1/hipp.Plx)
hipp['M_v'] = (hipp.Vmag - 5 * numpy.log10(hipp.D) + 5)

hipp
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.scatter(hipp['BV'], hipp['M_v'], s=1, c='navy',alpha=0.1)

ax.set_title('Color-Absolute Magnitude Diagram')
ax.title.set_fontsize(20)

ax.set_box_aspect(1)

ax.set_xlabel(r'B-V (mag)')
ax.set_ylabel(r'M$_V$ (mag)')

ax.set_xlim(-0.5, 2)
ax.set_ylim(25, 10)
```
